## Messages

```python
from langchain_core.messages import HumanMessage, SystemMessage
```

```python
messages=[
    SystemMessage(content='Translate the following from English to German'),
    HumanMessage(content='Hi, how are you?')
]
```

```python
response = llm.invoke(messages)
```

```python
response
```

## Output Parsing - StrOutputParser()

```python
from langchain_core.output_parsers import StrOutputParser
```

```python
parser = StrOutputParser()
```

```python
parser.invoke(response)
```

## Chaining Components - with LCEL

```python
chain=llm|parser
```

```python
chain.invoke(messages)
```

### Prompt Template

```python
from langchain_core.prompts import ChatPromptTemplate
```

```python
prompt_template='Translate the following into {language}:'
```

```python
prompt=ChatPromptTemplate.from_messages([
    ('system',prompt_template),
    ('user','{text}')
])
```

```python
result=prompt.invoke({
    'language':'German',
    'text':'Hi'
})
```

```python
result.to_messages()
```

### Chaining: Prompts + LLM + Parser

```python
chain=prompt|llm|parser
```

```python
chain.invoke({
    'language':'German',
    'text':'Hi'
})
```
